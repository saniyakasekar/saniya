```python
l1 = [1,2,3,4]
l2 = [3,4,5,6]
#add this two list element
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]

# Add elements element-wise using list comprehension
result = [x + y for x, y in zip(l1, l2)]

# Print the result
print(result)
```

```python
#Find maximum element combining two list l1 and l2
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]

# Combine the lists
combined_list = l1 + l2

# Find the maximum element
max_element = max(combined_list)

# Print the maximum element
print(max_element)
```

```python
# find the common element from list l1 and l2
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]

# Find common elements using set intersection
common_elements = list(set(l1) & set(l2))

# Print the common elements
print("Common elements:", common_elements)
```

```python
d1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
d2 = {'a': 7, 'f': 2, 'g': 55, 'h': "101"}

# Find common keys
common_keys = d1.keys() & d2.keys()  # Using set intersection

# Find common values
common_values = set(d1.values()) & set(d2.values())  # Using set intersection

# Print the results
print("Common keys:", common_keys)
print("Common values:", common_values)
```

```python
# replace the value of d1 to there negative value
for key,in d1 :
  d1[key]=-d1[key]
  print(d1)
```

```python
#replace key to value and value to key in d2
d2 = {'a': 7, 'f': 2, 'g': 55, 'h': "101"}

# Replace keys with values and values with keys
new_d2 = {value: key for key, value in d2.items()}

# Print the updated dictionary
print(new_d2)
```
